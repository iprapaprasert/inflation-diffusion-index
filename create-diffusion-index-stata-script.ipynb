```python
stata_vars = []
for indexes in vars.split():
    stata_var = "var"
    for index in indexes:
        if index != "0":
            stata_var = "".join([stata_var, index])
        else:
            break
    stata_vars.append(stata_var)

stata_vars
```

```python
# create stata script
import re
scripts = []
for stata_var in stata_vars:
    stata_script = f"nl ({stata_var}"
    children = [index for index in stata_vars if re.search(f"{stata_var}\d$", index)]
    if children:
        for i, child in enumerate(children):
            character = chr(ord('`') + i + 1)
            child_script = f"{{ln{character}}}*{child}"
            if i == 0:
                stata_script = " = ".join([stata_script, child_script])
            else:
                stata_script = " + ".join([stata_script, child_script])
        scripts.append(f"{stata_script}), nolog")

scripts
```
